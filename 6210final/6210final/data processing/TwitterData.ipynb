```python
movie=pd.read_csv("movies.csv",encoding="utf-8")
```

```python
movie
```

```python
posts=[]
users=[]
for index,row in movie.iterrows():
    if index == 50:
        break
    name=row['title']
    name=urllib.parse.quote(name,encoding="utf-8")
    a=api.GetSearch(raw_query="q="+name+"&count=100")
    
    for post in a:
        post=json.loads(str(post))
        postInfo=dict.fromkeys(['movieid',"postID","time","userID","post",'likes',"datasource",'hashtag','retweets'])
        postInfo["movieid"]=row['id']
        postInfo['datasource']="twitter"
        postInfo['likes']=post.get('favorite_count',0)
        postInfo["postID"]=post["id_str"]
        postInfo['hashtag']="#" + row['title']
        postInfo["time"]=time.strftime('%Y-%m-%d %H:%M:%S', \
                                       time.strptime(post["created_at"],'%a %b %d %H:%M:%S +0000 %Y'))
        postInfo["userID"]=post["user"]["id_str"]
        postInfo["post"]=post["text"]
        postInfo['retweets']=post.get('retweet_count',0)
        userInfo=dict.fromkeys(["username","userid","location","follower","createtime",'datasource'])
        userInfo["username"]=post["user"].get("name", None)
        userInfo["userid"]=post["user"]["id_str"]
        userInfo["datasource"]="twitter"
        userInfo["location"]=post["user"].get("location",None)
        userInfo["follower"]=post["user"].get("followers_count",0)
        userInfo["createtime"]=time.strftime('%Y-%m-%d %H:%M:%S', \
                                       time.strptime(post["user"]["created_at"],'%a %b %d %H:%M:%S +0000 %Y'))
        posts.append(postInfo)
        users.append(userInfo)
```

```python
posts
```

```python
posts = pd.DataFrame(posts)
users = pd.DataFrame(users)
users.drop_duplicates(subset=['userid'], inplace=True)
posts.drop_duplicates(subset=['postID'], inplace=True)
```

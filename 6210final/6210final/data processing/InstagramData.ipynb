```python
dataset = list()
i = 0
for index,row in movie.iterrows():
    if index == 50:
        break
    print(index)
    name = row['title'].replace(" ","").replace('-','').replace(':','').replace(',','').replace("'","").replace(".","").replace("!","").replace("&","and")
    posts=L.get_hashtag_posts(name)
    j = 0
    for post in posts:
        if j < 50:
            PostId = i
            #Hashtag = L.download_post(post, target='#'+moviee)
            Hashtag = '#'+name
            Caption = post.caption
            Profile = post.profile
            
            Likes = post.likes
            Comment = post.comments
            Time = post.date
            User_id = post.owner_id
            
            dataset.append((row['id'],PostId, Hashtag, Caption, Profile,Likes, Comment, Time, User_id,"instagram"))
            j = j+1
            i = i+1
        else:
            break
```

```python
Post = pd.DataFrame(dataset, columns = ['movie_id','Post_Id', 'Hashtag', 'Caption', 'Profile','Likes','Comments', 'Time','User_id','datasource'])
Post
```

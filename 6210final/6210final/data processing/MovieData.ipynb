#### Download the first 100 pages of top rated movies through TMDB API

```python
# First initiate two empty list for storing the raw data
rawTopMoviesJsonData = []
topMovies = []

getTopRateMoviesRequest = 'https://api.themoviedb.org/3/movie/top_rated?api_key=' + api_key + '&language=en-US&page='

for page in range(1, 41):
    movieData = requests.get(getTopRateMoviesRequest + str(page))
    rawTopMoviesJsonData.append(movieData.json())
    topMovies += movieData.json()['results']
```

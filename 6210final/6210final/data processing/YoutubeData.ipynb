3. Use the movie name to search on the Youtube and get the first five videos which are related to the movie.

```python
movieid=[]
movieName=[]
kind=[]
videoid=[]
publishedAt=[]
channelId=[]
title=[]
des=[]
parameters = {"part": "snippet",
              "maxResults":5 ,
              "pageToken": "",
              "q":"",
              "key": api_key,
              
              }
url = "https://www.googleapis.com/youtube/v3/search"
for n in range(0,50):
    name=df['title'][n]
    parameters["q"] =name
    page = requests.request(method="get", url=url, params=parameters)
    j_results = json.loads(page.text)
    item=j_results.get("items")
    print(name)
    for i in range(0,5):
        if(item[i] is None):
            next
        else:
            movieid.append(mid[n])
            movieName.append(name)
            kind.append(item[i].get("id").get("kind"))
            videoid.append(item[i].get("id").get("videoId"))
            publishedAt.append(item[i].get("snippet").get("publishedAt"))
            channelId.append(item[i].get("snippet").get("channelId"))
            title.append(item[i].get("snippet").get("title"))
            des.append(item[i].get("snippet").get("description"))
```

```python
tags1=[]
regionRes1=[]
like1=[]
dislike1=[]
favorite1=[]
comment1=[]
```

```python
parameters = {"part": "snippet,contentDetails,statistics",
              "maxResults":50 ,
              "pageToken": "",
              "id":"",
              "key": api_key,
              
              }
url = "https://www.googleapis.com/youtube/v3/videos"
```

```python
for i in range(0,250):
    print(i)
    parameters["id"] = videoid[i]
    page = requests.request(method="get", url=url, params=parameters)
    video_results = json.loads(page.text)
    if(video_results.get("items") is None):
            tags1.append("")
            regionRes1.append("")
            like1.append("")
            dislike1.append("")
            favorite1.append("")
            comment1.append("")
    elif(video_results.get("items")==[]):
            tags1.append("")
            regionRes1.append("")
            like1.append("")
            dislike1.append("")
            favorite1.append("")
            comment1.append("")
    else:
        if(video_results["items"][0]["snippet"].get("tags")==[]):
            tags1.append('')
        else:
            tags1.append(video_results["items"][0]["snippet"].get("tags"))
        regionRes1.append(video_results["items"][0]["contentDetails"].get("regionRestriction"))
        like1.append(video_results["items"][0]["statistics"].get("likeCount"))
        dislike1.append(video_results["items"][0]["statistics"].get("dislikeCount"))
        favorite1.append(video_results["items"][0]["statistics"].get("favoriteCount"))
        comment1.append(video_results["items"][0]["statistics"].get("commentCount"))
```

5. Get the unique author id and pass it to the channels of the youtube api and we can get the author info.

```python
authorId=movie_comments_df['UserId'].unique()
l=len(authorId)
```

```python
authorParameters = {"part": "snippet,statistics",
              "id": "UCoTDiXx0P9Q2XB8_xjYBQbw",
              "key": api_key,
              }
url ='https://www.googleapis.com/youtube/v3/channels'
viewerId=[]
viewerTitle=[]
viewerDes=[]
viewerCountry=[]
viewerView=[]
viewerComment=[]
viewerSub=[]
viewerVideo=[]
created=[]

```

```python
for i in range(0,l):
    aid=authorId[i]
    authorParameters["id"] = aid
    page = requests.request(method="get", url=url, params=authorParameters)
    viewer_results = json.loads(page.text)
    viewerTitle.append(viewer_results['items'][0]['snippet']['title'])
    viewerDes.append(viewer_results['items'][0]['snippet']['description'])
    viewerCountry.append(viewer_results['items'][0]['snippet'].get('country'))
    viewerView.append(viewer_results['items'][0]['statistics']['viewCount'])
    viewerComment.append(viewer_results['items'][0]['statistics']['commentCount'])
    viewerSub.append(viewer_results['items'][0]['statistics']['subscriberCount'])
    viewerVideo.append(viewer_results['items'][0]['statistics']['videoCount'])
    created.append(viewer_results['items'][0]['snippet']['publishedAt'])
    viewerId.append(aid)
```
